```python
extractedText=[]

def extract_text_from_pdf(path):
    print(path)
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

for pdf_file in uploaded.keys():
    print(pdf_file)
    pdf_text = extract_text_from_pdf(pdf_file)
    extractedText.append(pdf_text)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load tokenizer and model
model_name = "Qwen/Qwen1.5-0.5B"  # You can change to "Qwen/Qwen1.5-1.8B" for a larger model
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, trust_remote_code=True).eval()
```

```python
!pip install nltk
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
def chunk_by_sentences(text, max_words=150):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = ""

    for sentence in sentences:
        if len((current_chunk + " " + sentence).split()) <= max_words:
            current_chunk += " " + sentence
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence

    if current_chunk:
        chunks.append(current_chunk.strip())

    return chunks
```

```python
chunked_texts = []

for pdf_text in extractedText:
    chunks = chunk_by_sentences(pdf_text, max_words=150)
    chunked_texts.append(chunks)
```

```python
print(len(chunked_texts[4]))
```

```python
# Prepare training data in prompt-response format
formatted_data = []

for pdf_chunks in chunked_texts:
    for chunk in pdf_chunks:
        prompt = "Read and learn from the following text:\n"
        response = chunk  # The chunk itself will be the "response"
        formatted_data.append({"prompt": prompt, "response": response})
```

```python
def tokenize_example(example):
    return tokenizer(example["prompt"] + example["response"], truncation=True, padding="max_length", max_length=512)

# Tokenize the formatted data
tokenized_data = list(map(tokenize_example, formatted_data))
```

```python
!pip install datasets
```

```python
# Add 'labels' to each item in the tokenized data
for item in tokenized_data:
    item["labels"] = item["input_ids"].copy()

# Now create the dataset
from datasets import Dataset
dataset = Dataset.from_list(tokenized_data)

```

```python
print(dataset)
```

```python
from transformers import Trainer, TrainingArguments

# Define training arguments
training_args = TrainingArguments(
    output_dir="./qwen-finetuned",  # Output directory for the model
    per_device_train_batch_size=4,  # Batch size per device
    num_train_epochs=3,  # Number of epochs to train
    logging_dir='./logs',  # Directory for logs
    save_steps=500,  # Save checkpoint every 500 steps
    logging_steps=100,  # Log training info every 100 steps
    fp16=True,  # Use mixed precision training
    learning_rate=5e-5,  # Learning rate
)

# Initialize Trainer
trainer = Trainer(
    model=model,  # Pre-trained model
    args=training_args,  # Training arguments
    train_dataset=dataset,  # Our tokenized dataset
)

# Start fine-tuning
trainer.train()
```

```python
model.save_pretrained("qwen-finetuned")
tokenizer.save_pretrained("qwen-finetuned")
```
